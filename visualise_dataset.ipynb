# Fashion-MNIST Dataset
A quick look at the Fashion-MNIST dataset used in this coursework. Note: Please do not submit this file as part of your assignment.

From the authors: "`Fashion-MNIST` is a dataset of [Zalando](https://jobs.zalando.com/tech/)'s article imagesâ€”consisting of a training set of 60,000 examples and a test set of 10,000 examples. Each example is a 28x28 grayscale image, associated with a label from 10 classes."

Each image is a single clothing item from one of these 10 categories:

| Label | Class |
|-------|-------|
| 0 | T-shirt/top |
| 1 | Trouser |
| 2 | Pullover |
| 3 | Dress |
| 4 | Coat |
| 5 | Sandal |
| 6 | Shirt |
| 7 | Sneaker |
| 8 | Bag |
| 9 | Ankle boot |

## Load the dataset

```python
import matplotlib.pyplot as plt

from utils import load_training_data
# Load dataset
dataset = load_training_data()

# Class names
classes = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
           'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

print(f"Dataset size: {len(dataset)} images")
print(f"Image shape: 28x28 grayscale")
print(f"Classes: {len(classes)}")
```

## Visualise Sample Images

```python
# Display sample images
fig, axes = plt.subplots(2, 5, figsize=(12, 5))
fig.suptitle('Fashion-MNIST Sample Images', fontsize=16)

for i, ax in enumerate(axes.flat):
    image, label = dataset[i]
    ax.imshow(image, cmap='gray')
    ax.set_title(classes[label])
    ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Show one image in detail
image, label = dataset[0]

plt.figure(figsize=(4, 4))
plt.imshow(image, cmap='gray')
plt.title(f'Label: {label} ({classes[label]})')
plt.axis('off')
plt.show()

print(f"Image size: {image.size}")
print(f"Label: {label}")
```
